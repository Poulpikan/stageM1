```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv1D, MaxPooling1D, UpSampling1D
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Cropping1D
# Variables
window_size = 100
X_all = []
test = 0

# Chargement des séquences
f = FastA("../data/Fasta/TriTrypDB-68_LmajorFriedlin_Genome.fasta")
for maseq in f:
    test += 1
    if test <= 5:
        print(maseq.name)
        seq = maseq.sequence.upper()
        s = DNA(seq)
        s.window = window_size

        df_karlin = pd.read_csv(f"../data/karlin/{maseq.name}_karlin_{window_size}.csv")
        karlin = df_karlin['Karlin'].tolist()
        entropy = s.get_entropy(window_size)
        karlin_scalar = [karlin[i] for i in range(0, len(karlin), window_size)]
        entropy_scalar = [entropy[i] for i in range(0, len(entropy), window_size)]

        y_vals = [0]
        for base in seq:
            y = y_vals[-1]
            if base in ('A', 'C'):
                y += 1
            elif base in ('G', 'T'):
                y -= 1
            y_vals.append(y)
        at = []
        gc = []
        for i in range(0, len(seq) - window_size + 1, window_size):
            window = seq[i:i + window_size]
            g = window.count("G")
            c = window.count("C")
            a = window.count("A")
            t = window.count("T")
        
            gc.append((g + c) / (a + t + g + c))
            at.append((a + t) / (a + t + g + c))
        at.append((a + t) / (a + t + g + c))
        entropy_scalar = at
        window_means = [np.mean(y_vals[i:i+window_size]) for i in range(0, len(y_vals), window_size)]
        min_len = min(len(karlin_scalar), len(window_means))
        X_chr = np.column_stack((
            karlin_scalar[:min_len],
            window_means[:min_len],
    #        entropy_scalar[:min_len]
        ))
        X_all.append(X_chr)


# Normalisation et padding par chromosome
X_scaled_all = []
scaler = MinMaxScaler()
for X_chr in X_all:
    X_scaled_chr = scaler.fit_transform(X_chr)
    X_scaled_all.append(X_scaled_chr)

# Padding des séquences
X_padded = pad_sequences(X_scaled_all, padding='post', dtype='float32')

# Construction du modèle Conv1D
input_layer = Input(shape=(X_padded.shape[1], 2))

# Encodeur
x = Conv1D(2, 10, activation='relu', padding='same')(input_layer)
x = MaxPooling1D(2, padding='same')(x)
x = Conv1D(1, 10, activation='relu', padding='same')(x)
encoded = MaxPooling1D(2, padding='same')(x)

# Decodeur
x = Conv1D(1, 10, activation='relu', padding='same')(encoded)
x = UpSampling1D(2)(x)
x = Conv1D(2, 10, activation='relu', padding='same')(x)

x = UpSampling1D(2)(x)
decoded = Conv1D(2, 2, activation='sigmoid', padding='same')(x)

# Ajustement de la taille de sortie
output_len = X_padded.shape[1]
decoded = Cropping1D((0, decoded.shape[1] - output_len))(decoded)

# Modèle final
autoencoder = Model(input_layer, decoded)

# Compilation
autoencoder = Model(input_layer, decoded)
autoencoder.compile(optimizer=Adam(learning_rate=0.001), loss='mse')
autoencoder.summary()

# Entraînement
history = autoencoder.fit(X_padded, X_padded,
                          epochs=50,
                          batch_size=1,
                          validation_split=0.2,
                          shuffle=True)

# Visualisation de la perte
plt.plot(history.history['loss'], label='Perte (train)')
plt.plot(history.history['val_loss'], label='Perte (validation)')
plt.xlabel('Époques')
plt.ylabel('MSE')
plt.legend()
plt.title("Évolution de la perte")
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv1D, MaxPooling1D, UpSampling1D, Cropping1D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

window_size = 100
X_all = []
scalers = []
test = 0

f = FastA("../data/Fasta/TriTrypDB-68_LmajorFriedlin_Genome.fasta")

for maseq in f:
    test += 1
    if test = 7:
        print(f"Chargement : {maseq.name}")
        seq = maseq.sequence.upper()
        s = DNA(seq)
        s.window = window_size

        df_karlin = pd.read_csv(f"../data/karlin/{maseq.name}_karlin_{window_size}.csv")
        karlin = df_karlin['Karlin'].tolist()
        entropy = s.get_entropy(window_size)
        karlin_scalar = [karlin[i] for i in range(0, len(karlin), window_size)]
        entropy_scalar = [entropy[i] for i in range(0, len(entropy), window_size)]

        y_vals = [0]
        for base in seq:
            y = y_vals[-1]
            if base in ('A', 'C'):
                y += 1
            elif base in ('G', 'T'):
                y -= 1
            y_vals.append(y)

        window_means = [np.mean(y_vals[i:i+window_size]) for i in range(0, len(y_vals), window_size)]
        min_len = min(len(karlin_scalar), len(window_means), len(entropy_scalar))
        X_chr = np.column_stack((
            karlin_scalar[:min_len],
            window_means[:min_len],
            entropy_scalar[:min_len]
        ))

        # Normalisation par chromosome
        scaler = MinMaxScaler()
        X_chr_scaled = scaler.fit_transform(X_chr)
        scalers.append(scaler)
        X_all.append(X_chr_scaled)

# --- Séparation : 4 premiers pour entraînement, 5e pour test ---
X_train = X_all[:4]
X_test = X_all[4]

# Padding
X_train_padded = pad_sequences(X_train, padding='post', dtype='float32')
X_test_padded = pad_sequences([X_test], maxlen=X_train_padded.shape[1], padding='post', dtype='float32')

# Modèle
input_layer = Input(shape=(X_train_padded.shape[1], 3))
x = Conv1D(16, 3, activation='relu', padding='same')(input_layer)
x = MaxPooling1D(2, padding='same')(x)
x = Conv1D(8, 3, activation='relu', padding='same')(x)
encoded = MaxPooling1D(2, padding='same')(x)

x = Conv1D(8, 3, activation='relu', padding='same')(encoded)
x = UpSampling1D(2)(x)
x = Conv1D(16, 3, activation='relu', padding='same')(x)
x = UpSampling1D(2)(x)
decoded = Conv1D(3, 3, activation='sigmoid', padding='same')(x)

# Ajustement longueur de sortie
output_len = X_train_padded.shape[1]
decoded = Cropping1D((0, decoded.shape[1] - output_len))(decoded)

autoencoder = Model(input_layer, decoded)
autoencoder.compile(optimizer=Adam(learning_rate=0.001), loss='mse')
autoencoder.summary()

# Entraînement
history = autoencoder.fit(X_train_padded, X_train_padded,
                          epochs=50,
                          batch_size=1,
                           validation_data=(X_test_padded, X_test_padded),
                          shuffle=True)

# Visualisation de la perte
plt.plot(history.history['loss'], label='Perte (train)')
plt.plot(history.history['val_loss'], label='Perte (validation)')
plt.xlabel('Époques')
plt.ylabel('MSE')
plt.legend()
plt.title("Évolution de la perte")
plt.show()

# --- Prédiction sur le chromosome 5 ---
X_test_pred = autoencoder.predict(X_test_padded)

# Affichage d'une comparaison sur la 1ʳᵉ feature (ex: Karlin)
plt.figure(figsize=(12, 4))
plt.plot(X_test_padded[0][:, 1], label='Karlin original')
plt.plot(X_test_pred[0][:, 1], label='Karlin reconstruit', linestyle='--')
plt.legend()
plt.title("Reconstruction - Chromosome 5")
plt.show()
```
